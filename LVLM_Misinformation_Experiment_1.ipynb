```python
# Path to the ZIP file
zip_path = "/content/drive/My Drive/MOCHEG/latest_dataset.zip"

# Unzip into the target directory
unzip_dir = "/content/drive/My Drive/MOCHEG/unzipped/"
os.makedirs(unzip_dir, exist_ok=True)

# Unzipping the file
!unzip -q "{zip_path}" -d "{unzip_dir}"

# Verify the extracted files
!ls "{unzip_dir}"
```

```python
# Define paths
tar_file_path = "/content/drive/My Drive/MOCHEG/latest_dataset.tar.gz"  # Path to the downloaded .tar.gz file
extract_to_path = "/content/drive/My Drive/MOCHEG/extracted/"  # Path to extract files

# Create the extraction directory if it doesn't exist
import os
os.makedirs(extract_to_path, exist_ok=True)

# Extract the .tar.gz file
!tar -xvzf "{tar_file_path}" -C "{extract_to_path}"

# Check if extraction was successful
if os.path.exists(extract_to_path):
    print("Extraction completed successfully!")

    # Delete the original .tar.gz file
    os.remove(tar_file_path)
    print(f"Deleted the tar.gz file: {tar_file_path}")
else:
    print("Extraction failed. The original file will not be deleted.")
```
